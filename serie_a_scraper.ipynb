```python
seasons = [(f"{year}-{year+1}") for year in range(2022, 2024)]
url = "https://fbref.com/en/comps/11/"
final = "/Serie-A-Stats"
section = "matchlogs/all_comps/shooting/"
all_matches = []
```

```python
import random

#url + season + final is the final url for each season

for season in seasons:
    print("scraping season:", season)   
    headers = {'User-Agent': random.choice(user_agents)}
    page = requests.get(url + season + final, headers=headers).text
    soup = BeautifulSoup(page)
    teams_table = soup.select("table.stats_table")[0]
    links = teams_table.select("a")
    team_links = [link.get("href") for link in links if "/en/squads" in link.get("href")]
    for team in team_links:
        
        team_name = team.split("/")[-1].replace("Stats", "").replace("-", "")
        print("scraping team:", team_name)
        team_url = base_url + team

        headers = {'User-Agent': random.choice(user_agents)}
        data = requests.get(team_url, headers=headers).text
        data_io = io.StringIO(data)
        matches = pd.read_html(data_io, attrs={"id": "matchlogs_for"})[0]
        #need to scrape also the shooting table
        shooting_url = team_url.split("/")
        shooting_url[-1] = section + shooting_url[-1]
        shooting_url = "/".join(shooting_url)

        headers = {'User-Agent': random.choice(user_agents)}
        data = requests.get(shooting_url, headers=headers).text
        data_io = io.StringIO(data)
        shooting = pd.read_html(data_io, attrs={"id": "matchlogs_for"})[0]
        shooting.columns = shooting.columns.droplevel()

        try:
            team_df = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")

        except ValueError:
            continue
        
        team_df = team_df[team_df["Comp"] == "Serie A"]
        team_df["Team"] = team_name
        team_df["Season"] = season.split("-")[0]
        all_matches.append(team_df)
        time.sleep(random.randint(40, 60)) #sleep to prevent error 429 (too many requests)
    
```

```python
matches_df = pd.concat(all_matches)
```

```python
matches_df.columns = [c.lower() for c in matches_df.columns]
```
